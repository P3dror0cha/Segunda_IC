```python
import requests
import pandas as pd
from io import StringIO
```

```python
url = "https://www.ebi.ac.uk/metagenomics/api/latest/studies/MGYS00006848/samples?format=csv"
def download_studies_metadata(url):
    resposta = requests.get(url)
    if resposta.status_code == 200:
        df = pd.read_csv(StringIO(resposta.text))
        print("Esta é a resposta!")
        print(df.head)
        df.to_csv("studies.csv")
    else:
        print(f"Deu problema no status {resposta.status_code}")

    return df

df = download_studies_metadata(url)
```

```python
df_lista_ids_sample = df["accession"].to_list()
print(df_lista_ids_sample)
```

```python

def download_samples_metadata(sample_id):
    
    for sample in sample_id:
        url_samples = f"https://www.ebi.ac.uk/metagenomics/api/latest/samples/{sample}/runs?format=csv"
        resposta = requests.get(url_samples)

        if resposta.status_code == 200:
            df_sample = pd.read_csv(StringIO(resposta.text))
            print("Esta é a resposta!")
            print(df_sample.head)
            df_sample.to_csv(f"sample_{sample}_metadata.csv")
        else:
            print(f"Deu problema no status {resposta.status_code}")

    return df_sample

df_2 = download_samples_metadata(df_lista_ids_sample)
```
