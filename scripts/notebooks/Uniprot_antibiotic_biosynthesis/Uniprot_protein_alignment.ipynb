### Uniprot alignment
Here we take the output from the previous notebook (uniprot_protein_download.ipynb) and uses the dataframe as the input for the manipulations here. 

```python
import pandas as pd

df = pd.read_csv("/home/pedro/antismash/BIG-SCAPE/uniprot_table.csv")
df
```

```python
df["kingdom"].value_counts()
```

```python
df_filtered = df[df["kingdom"] == "Bacteria"]
df_filtered
```

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

records = [
    SeqRecord(
        Seq(seq),
        id=uid,
        description=""
    )
    for uid, seq in zip(df_filtered["accession"], df_filtered["sequence"])
]

SeqIO.write(records, "uniprot_antibiotc_biosynthesis_protein.faa", "fasta")
```
